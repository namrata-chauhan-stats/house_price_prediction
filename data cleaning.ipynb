```python
# to convert price column into Crores for all the datasets
def price_chng(x):
  if type(x) == float:
    return x

  else:
    if x[1]=='Lac':
      return round(float(x[0])/100,2)
    else:
      return round(float(x[0]),2)
```

```python
# Checking using a list
print(price_chng(['90.5', 'Lac']))
```

```python
df['price'].str.split(' ').apply(price_chng) # final conversion of the price col
```

```python
df['price']=df['price'].str.split(' ').apply(price_chng)
```
